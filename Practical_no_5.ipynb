sound signal**

```python
#importing the libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# now using define function here for the image 
def compare_image(image1, image2):
  plt.figure(figsize=(9,9))
  plt.subplot(1,2,1)
  plt.imshow(image1)
  plt.title('Orignal')
  plt.axis('off')

  plt.subplot(1,2,2)
  plt.imshow(image2)
  plt.title('Modified')
  plt.axis('off')
```

```python
# now to make the image read by python we will use the following code here 
img = cv2.imread('/content/cycle.jpg')

```

```python
#by default python read the image in BGR format, to see the original format that is RGB will use this code
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
compare_image(img2,img)
```

```python
#now this code is used to perform convolution operation on the image 
# which actually use matrix for it, (dont go into very deep it is just for understanding) 
kernel = np.ones((10,10),np.float32)/100
cnv = cv2.filter2D(img2, -1, kernel)
compare_image(img2,cnv)
```

**Now From here will start the smoothning of image, where will use 1.Averaging(bluring the image) and 2.Median Filtering(removing noise from the image)**



```python
# 1.Average Filtering
blur = cv2.blur(img2,(10,10))
compare_image(img2,blur)
```

```python
# 3. Median Filtering 

# now first we introduce some random noise for importing that we use this particular code 
#reason of introducing noise so we can apply the median filter

from skimage.util import random_noise


```

```python
## for adding noise we will use this code 
noise_img = random_noise(img2, mode='s&p',amount=0.3)
noise_img = np.array(255*noise_img, dtype = 'uint8')


```

```python
## applying median filter
median = cv2.medianBlur(noise_img,5)
compare_image(noise_img,median)
```
