**Write the program to implement various morphological image processing techniques. **

"there are several methods in marphological processing but we will use 1. erosion  and   2.Dilation" 

```python
#importing the libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# now using define function here for the image 
def compare_image(image1, image2):
  plt.figure(figsize=(9,9))
  plt.subplot(1,2,1)
  plt.imshow(image1)
  plt.title('Orignal')
  plt.axis('off')

  plt.subplot(1,2,2)
  plt.imshow(image2)
  plt.title('Modified')
  plt.axis('off')
```

```python
# now to make the image read by python we will use the following code here 
img = cv2.imread('/content/forest.jpg',0)
```

```python
#by default python read the image in BGR format, to see the original format that is RGB will use this code

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# convolution operation 
kernel = np.ones((5,5),np.uint8)

```

```python
#1. erosion code
erosion = cv2.erode(img,kernel,iterations=1)
compare_image(img,erosion)
```

```python
# 2. Dilation code
dia = cv2.dilate(img,kernel,iterations=1)
compare_image(img,dia)
```
